# Find temperature in each year

```python
frames = [data_Taipei_2013,data_Taipei_2014, data_Taipei_2015, data_Taipei_2016, data_Taipei_2017, data_Taipei_2018, data_Taipei_2019]
data = pd.concat(frames)
```

```python
data.info()
```

```python
data.groupby(['Year','Month'])['Temperature'].max()
```

```python
fig, ax2 = plt.subplots(figsize=(15,7))
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("Mean temperature in each year")
data_month.groupby(['Month','Year'])['Temperature'].mean().unstack().plot(ax=ax2)
```

```python
fig, ax2 = plt.subplots(figsize=(15,7))
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("Min temperature in each year")
data_month.groupby(['Month','Year'])['Temperature'].min().unstack().plot(ax=ax2)
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
#plt.xlabel("")
plt.ylabel("Temperature")
plt.title("mean temperature in each year")
# use unstack()
data_Year.groupby('Year')['Temperature'].mean().plot(ax=ax)
```
